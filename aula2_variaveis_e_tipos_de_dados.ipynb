```python
type(10.0)
```

Em Python (e quase todas as linguagens), usamos o padrão americano. Se você usar vírgula, o Python vai achar que você está criando uma lista ou tupla

```python
x = 7,5
type(x)
```

```python
#Texto/String(str): Sequência de texto, como: "Olá", "123". Representadas por aspas duplas ("") ou aspas simples (' ')
type("hello")
```

```python
#Lógico (bool): Representam True(verdadeiro) e False(falso)
type(True)
```

```python
print(5 + 2.0) #O Python é inteligente. Se você somar um número inteiro com um float, o resultado será sempre um float.
```

```python
# Se um número estiver entre aspas, será considerado texto e a adição virará concatenção
x = "123"
y = "45"
print(x + y)
```

```python
print('12' + 20)
```

## Variáveis em Python

```python
mensagem = 'Hello, world!'
print(mensagem)
```

```python
#Isso gerará um erro por conta do espaço
meu nome = 'Maria'
print(meu nome)
```

```python
# Idealmente usar snake_case!
meu_nome = 'Maria'
print(meu_nome)
```

```python
idade = 29
print('A minha idade é:', idade)
```

```python
IDADE = 30
print(idade, IDADE)
```

```python
user = "mariaclara"
```

```python
print('Hello, {}!'.format(user))
```

```python
print(f'Hello, {user}!')
```

```python
print('Hello, {}! Você tem {} anos'.format(user, idade))
```

```python
print(f'Hello, {user}! Você tem {idade} anos')
```
