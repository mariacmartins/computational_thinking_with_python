# Aula de Match/Case

```python
dia = 1

if dia == 1:
    print("Domingo")
elif dia == 2:
    print("Segunda")
elif dia == 3:
    print("Terça")
elif dia == 4:
    print("Quarta")
elif dia == 5:
    print("Quinta")
elif dia == 6:
    print("Sexta")
elif dia == 7:
    print("Sábado")
else:
    print("Dia Inválido")
```

```python
dia = 1

match dia:
    case 1:
        print("Domingo")
    case 2:
        print("Segunda")
    case 3:
        print("Terça")
    case 4:
        print("Quarta")
    case 5:
        print("Quinta")
    case 6:
        print("Sexta")
    case 7:
        print("Sábado")
    case _:
        print("Dia Inválido")
```

O if/elif é uma estrutura genérica de controle de fluxo que pode testar qualquer condição (ex: dia > 5). Já o match/case deixa claro que você está fazendo uma comparação de padrão (Pattern Matching) em cima de uma única variável. É visualmente mais direto.

No match/case, o uso do _ (underscore) é o padrão para o "caso padrão" (default). Na Engenharia de Software, isso é uma convenção comum que indica: "se não cair em nenhum dos casos acima, faça isso". É mais elegante que um else solto.

Repare que no if/elif, você precisa escrever a variável dia == em todas as linhas. No match/case, você declara a variável uma única vez no topo. Isso reduz a chance de erros de digitação.

```python
#Exercício:
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

match operacao:
    case "+":
        print(f"Resultado: {num1 + num2}")
    case "-":
        print(f"Resultado: {num1 - num2}")
    case "*":
        print(f"Resultado: {num1 * num2}")
    case "/":
        print(f"Resultado: {num1 / num2}")
    case _:
        print("Operação não reconhecida.")
```

Redução de Repetição de Variáveis: No if/elif, você é obrigado a repetir o nome da variável e o operador de comparação (ex: dia ==) em todas as linhas. No match, você declara a variável uma única vez no topo, deixando o código mais "limpo" e menos poluído visualmente.

Minimização de Erros Humanos: Como você não precisa reescrever a variável em cada linha, elimina-se o risco de cometer erros de digitação (typos) no meio do bloco ou de usar o operador de atribuição = em vez do de comparação ==.

```python
codigo = int(input("Digite o código do setor (10-14): "))

match codigo:
    case 10:
        print("Setor: Alimentos")
    case 11:
        print("Setor: Bebidas")
    case 12:
        print("Setor: Limpeza")
    case 13:
        print("Setor: Higiene")
    case 14:
        print("Setor: Hortifruti")
    case _:
        print("Setor Desconhecido")
```

**Motor de Cálculo de Frete (E-commerce)**

Uma plataforma de e-commerce precisa calcular o valor do frete e o prazo de entrega com base na região do Brasil (representada por uma sigla de 2 letras).
Crie um programa que receba a regiao e defina as variáveis valor_frete (float) e prazo_dias (int).

"SP": R$ 10.00 | 2 dias

"RJ": R$ 15.00 | 3 dias

"MG": R$ 18.00 | 4 dias

Outros: R$ 50.00 | 20 dias

```python
regiao = input("Digite a sigla da região (SP, RJ, MG, etc): ").upper()

match regiao:
    case "SP":
        valor_frete = 10.0
        prazo_dias = 2
    case "RJ":
        valor_frete = 15.0
        prazo_dias = 3
    case "MG":
        valor_frete = 18.0
        prazo_dias = 4
    case _:
        valor_frete = 50.0
        prazo_dias = 20

print(f"Frete: R$ {valor_frete:.2f} | Entrega em {prazo_dias} dias.")
```

Objetivo: Criar um sistema que normalize leituras de sensores para a escala Celsius.Enunciado:Um software de monitoramento industrial recebe dois dados: um valor (float) e uma unidade (string).
Como o sistema central só entende graus Celsius, você deve converter a entrada seguindo estas regras:

Celsius ("c", "celsius"): O valor não muda.

Fahrenheit ("f", "fahrenheit"): Aplicar $C = (F - 32) / 1.8$.

Kelvin ("k", "kelvin"): Aplicar $C = K - 273.15$.

Qualquer outra entrada: Definir o resultado como None.

```python
# Entrada de dados
valor_entrada = float(input("Digite o valor da temperatura: "))
unidade = input("Digite a unidade (C, F ou K): ").strip().lower()

# Processamento com Match/Case
match unidade:
    case "c" | "celsius":
        resultado_celsius = valor_entrada
        status = "Sucesso"

    case "f" | "fahrenheit":
        resultado_celsius = (valor_entrada - 32) / 1.8
        status = "Sucesso"

    case "k" | "kelvin":
        resultado_celsius = valor_entrada - 273.15
        status = "Sucesso"

    case _:
        resultado_celsius = None
        status = "Erro: Unidade não reconhecida pelo sistema."

# Saída de dados (Lógica de apresentação)
if resultado_celsius is not None:
    print(f"Valor normalizado: {resultado_celsius:.2f}°C")
else:
    print(status)
```
