## Operadores Aritméticos

```python
# Soma
print(3 + 2)
```

```python
# Subtração: Também podemos usar variáveis pra isso
price = 20
discount = 4
final_price = price - discount
print(final_price)
```

```python
# Multiplicação
print(10 * 10)
```

```python
# Divisão: Se a operação de divisão for realizada com uma barra /, o valor resultante será um número de ponto flutuante (float).
print(5 / 2)
```

```python
# Divisão inteira: Se a operação de divisão for realizada com duas barras //, o valor resultante será um número inteiro (int).
# O valor será truncado, ou seja, as casas decimais serão ignoradas.

print(5//2)
```

```python
#Resto da Divisão: Se a operação de divisão for realizada com o operador %, teremos o resto de uma divisão entre números inteiros.
print(5%2)
```

Imagine que você está desenvolvendo um site e tem 43 produtos para exibir. Cada página do site suporta apenas 10 produtos.

Quantas páginas completas você terá? 43 // 10 = 4 páginas.

Quantos produtos sobrarão para a última página (incompleta)? 43 % 10 = 3 produtos.

```python
print(43/10)
print(43//10)
print(43%10)
```

Momento exercício: 5min
Escreva um programa que pergunte a quantidade de Km percorridos por um carro alugado e a quantidade de dias pelos quais ele foi alugado.

Calcule o preço a pagar, sabendo que o carro custa 80 reais por dia e 0.20 reais (20 centavos) por Km rodado.

```python
km = float(input('Quantos km foram percorridos? '))
days = int(input('Por quantos dias o carro foi alugado? '))

price = (days * 80) + (km * 0.20)
print(f'O preço a pagar é de {price} reais')
```

## Operadores Relacionais

```python
estoque = 50
pedido = 60

print(pedido > estoque)   # True (Estoque insuficiente)
print(pedido == estoque)  # False (Não é a quantidade exata)
```

```python
# Exemplo validando se o aluno passou de ano
nota = float(input("Digite a nota do aluno (0-100): "))
# Se fosse um programa real, poderia percorrer uma coluna de uma tabela com as notas dos alunos (aprenderemos isso em estruturas de repetição)

# 1. Verificação de Intervalo (Nota entre 0 e 10)
# Podemos usar a sintaxe encadeada do Python ou o operador 'and'
verifica_nota = (nota >= 60)

# 2. Verificação de Nota Máxima
nota_maxima = (nota == 10)

# Saída de resultados (Booleanos)
print(f"Passou? {verifica_nota}")
print(f"O aluno tirou a nota máxima? {nota_maxima}")
```

Exercício 5min:
Defina as variáveis stored_user = "admin" e stored_password = "1234".
Peça para o usuário digitar um login e uma senha (use as funções de input).
Use operadores relacionais para exibir True se o acesso for permitido (ou seja, se o usuário for admin) e False caso contrário.
Não é necessário usar condicionais if e else ainda, use apenas os operadores relacionais.

```python
# System registered data
stored_user = "admin"
stored_password = "1234"

# User input
input_user = input("Enter username: ")
input_password = input("Enter password: ")

is_user_correct = (input_user == stored_user)
is_password_correct = (input_password == stored_password)

# Output
print("-" * 20)
print(f"Username recognized? {is_user_correct}")
print(f"Password recognized? {is_password_correct}")
```

## Operadores Lógicos

```python
# Exemplo 1: Verificação de bônus salarial
salary = 3000
years_at_company = 5

# Regra: Salário menor que 5000 E mais de 3 anos de casa
has_bonus = (salary < 5000) and (years_at_company > 3)
print(f"Qualified for bonus? {has_bonus}") # True
```

```python
# Exemplo 2: Acesso ao sistema (Multi-fator)
has_keycard = False
has_fingerprint = True

# Regra: Entra se tiver o cartão OU a digital
can_enter = has_keycard or has_fingerprint
print(f"Can enter building? {can_enter}") # True
```

```python
# System registered data
stored_user = "admin"
stored_password = "1234"

# User input
input_user = input("Enter username: ")
input_password = input("Enter password: ")

# Authentication Logic
# Access is True only if BOTH username AND password are correct
is_access_granted = (input_user == stored_user) and (input_password == stored_password)


# Output
print("-" * 20)
print(f"Access granted? {is_access_granted}")
```

```python
# --- CONFIGURAÇÃO DOS FILTROS (Simulando o clique do usuário no site) ---
is_hotel = True
is_apart_hotel = False
is_near_airport = False

hotel = 'Hotel Colina das Montanhas'

# --- 1. LÓGICA DE FILTRO (Operadores Lógicos e Relacionais) ---
# A acomodação serve se: (É Hotel OU Apart) E (Perto do Aeroporto)
matches_search_criteria = (is_hotel or is_apart_hotel) and is_near_airport

# --- 2. EXIBIÇÃO ---
print(f"Exibir {hotel} para o usuário? {matches_search_criteria}")
```
