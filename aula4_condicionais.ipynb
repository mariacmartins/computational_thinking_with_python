## Estruturas condicionais

### if:

```python
# Entendendo a condição if
nota = float(input("Digite uma nota: "))

print(f'Nota do aluno no sistema: {nota}')

if nota >= 7:
  print('Aprovado(a)!')
```

### else:

```python
#Entendendo o else
nota = float(input("Digite uma nota: "))

print(f'Nota do aluno no sistema: {nota}')

if nota >= 7:
  print('Aprovado(a)!')

else:
  print('Não aprovado(a)')
```

### elif

```python
#Entendendo condicionais encadeadas (elif)
nota = float(input("Digite uma nota: "))

print(f'Nota do aluno no sistema: {nota}')

if nota >= 8:
  print('Aprovado(a) satisfatoriamente!')

elif nota == 7:
  print('Aprovado(a).')

else:
  print('Não aprovado(a).')
```

```python
# Nós também podemos combinar várias condições dentro dos blocos
#Exemplo de cálculo de média de notas
nota1 = float(input("Nota da primeira avaliação: "))
nota2 = float(input("Nota da segunda avaliação: "))
media = (nota1 + nota2) / 2

faltas = int(input("Quantidade de faltas no semestre: "))

if media >= 7 and faltas <= 10:
  print(f'Parabéns, você está aprovado!')

elif media >= 7 and faltas > 10:
  print(f'Você está reprovado! Faltas em excesso.')

else:
  print(f'Você está reprovado! Média insuficiente.')
```

### Comandos úteis para manipular strings (textos)

- startswith() — O que faz: Checa se a string começa com determinado texto.
- endswith() — O que faz: Checa se a string termina com determinado texto.
- strip() — O que faz: Remove espaços em branco inúteis no início e no fim da string.
- upper() —  Retorna a string em letras maiúsculas
- lower() —  Retorna a string em letras minúsculas

```python
raw_price = input("Valor do produto: ")  # ex: "120,50"

# trocando a vírgula pelo padrão americano (ponto)
adjusted_price = raw_price.replace(",", ".")

# Conversão para o tipo de dado correto
final_price = float(adjusted_price)

# Regra de negócio (Business Rule)
if final_price > 100:
    final_price *= 0.9  # Aplicando 10% de desconto
    print(f"Desconto aplicado! Valor final: R$ {final_price:.2f}")
else:
    print(f"Valor final: R$ {final_price:.2f}")
```

### Desafio: Validador de Checkout com Cupom Restrito

(10 min de resolução)

Você está desenvolvendo o sistema de um e-commerce. O programa deve processar o valor da compra e aplicar cupons de desconto baseados na categoria do produto.

Regras de Negócio:

Se o coupon_code for "TECH20" e a category for "eletronicos", o desconto é de 20%.

Se o coupon_code for "CASA10" e a category for "limpeza" ou "cozinha", o desconto é de 10%.

Caso o cupom não atenda aos requisitos ou o usuário não passou algum cupom, informe ao usuário.

```python
# Entrada de dados
total_purchase = float(input("Valor da compra (R$): "))
category = input("Categoria (eletronicos, limpeza, cozinha, etc): ").lower().strip()
coupon_code = input("Cupom de desconto: ").upper().strip()

# Variável para armazenar o percentual de desconto
discount_percent = 0.0

# Lógica das Regras de Negócio
if coupon_code == "TECH20" and category == "eletronicos":
    discount_percent = 0.20

elif coupon_code == "CASA10" and (category == "limpeza" or category == "cozinha"):
    discount_percent = 0.10

else:
    print("Aviso: Sem cupom ou cupom inválido para estas condições.")

# Cálculos
discount_value = total_purchase * discount_percent
final_price = total_purchase - discount_value

# Exibição do Resultado
print("-" * 30)
print(f"Categoria selecionada: {category}")
print(f"Desconto aplicado: R$ {discount_value}")
print(f"Total a pagar: R$ {final_price}")
```
