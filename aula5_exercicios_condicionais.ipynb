## Exercício: Sistema de Rank Up
Contexto:
Você foi contratado por uma organização de E-sports para automatizar o sistema de categorias dos jogadores. O objetivo é criar um script que analise o desempenho e entregue um veredito único.

Dados de Entrada (Inputs):

points: Um número inteiro representando a pontuação total.

win_rate: Um número decimal representando a taxa de vitória (0 a 100).

has_ban: Uma resposta textual ("sim" ou "nao") sobre punições ativas.

Regras de Negócio (Obrigatório seguir a ordem):

Bloqueio de Segurança: Se o jogador tiver um banimento ativo, o veredito deve ser "Bloqueado", independentemente de seus pontos ou vitórias.

Categoria Ouro: Para jogadores sem banimento, que possuam mais de 2500 pontos E taxa de vitória de 75% ou mais.

Categoria Prata: Para jogadores sem banimento, que possuam mais de 1000 pontos E taxa de vitória de 50% ou mais.

Categoria Bronze: Para todos os outros jogadores que não se encaixarem nos critérios acima.

```python
# 1. Coleta de Dados
points = int(input("Informe a pontuação: "))
win_rate = float(input("Informe a taxa de vitória (%): "))
has_ban = input("Possui banimento ativo? (sim/nao): ").lower().strip()

# 2. Processamento (Lógica de Caminhos Fechados)
if has_ban == "sim":
    rank = "Bloqueado"
else:
    # Este bloco só executa se o jogador NÃO estiver banido
    if points > 2500 and win_rate >= 75:
        rank = "Ouro"
    elif points > 1000 and win_rate >= 50:
        rank = "Prata"
    else:
        rank = "Bronze"

# 3. Exibição do Resultado
# Sem a inicialização global, garantimos que 'rank' existe em todos os caminhos acima
print(f"\nVeredito Final: {rank}")
```

## Exercício: Sistema simples de recomendação

Você está desenvolvendo o motor de sugestões de uma nova plataforma de streaming. O sistema deve recomendar um filme com base em duas variáveis: a idade do usuário (para garantir que ele veja algo apropriado) e o gênero favorito (Ação ou Drama).
Regras de Negócio:

Classificação Etária (Filtro Principal):

Se o usuário tiver menos de 14 anos, ele só pode receber recomendações de filmes da categoria "Infantojuvenil".

Se tiver 14 anos ou mais, o sistema deve checar o gênero preferido.

Lógica de Gêneros (Para Maiores de 14):

Se o gênero for "Ação", recomende: "Batman: O Cavaleiro das Trevas".

Se o gênero for "Drama", recomende: "O Lado Bom da Vida".

Lógica para Menores de 14:

Se o gênero for "Ação", recomende: "Os Incríveis".

Se o gênero for "Drama", recomende: "Extraordinário".

Tratamento de Erro: Se o usuário digitar um gênero que não seja "ação" ou "drama", o sistema deve dizer: "Gênero não disponível no catálogo".

```python
# Entrada de Dados
age = int(input("Qual sua idade? "))
genre = input("Qual seu gênero favorito (Ação ou Drama)? ").lower().strip()

# Variável para guardar a recomendação
movie_suggestion = ""

# 1. Verificação por idade (Categorização)
if age < 14:
    # Sub-lógica para menores
    if genre == "ação":
        movie_suggestion = "Os Incríveis"
    elif genre == "drama":
        movie_suggestion = "Extraordinário"
    else:
        movie_suggestion = "Gênero não encontrado"

else:
    # Sub-lógica para adultos/adolescentes
    if genre == "ação":
        movie_suggestion = "Batman: O Cavaleiro das Trevas"
    elif genre == "drama":
        movie_suggestion = "O Lado Bom da Vida"
    else:
        movie_suggestion = "Gênero não encontrado"

# Saída do Sistema
print("-" * 30)
if movie_suggestion == "Gênero não encontrado":
    print("Desculpe, ainda não temos esse gênero em nosso catálogo.")
else:
    print(f"Com base no seu perfil, sugerimos: {movie_suggestion}")
```

### Exercício:  Backend e E-commerce.

Este desafio combina manipulação de strings, lógica de prioridade e validação de estado (o usuário ser novo ou não).

Desafio: Sistema de Checkout Inteligente (Smart Cart)
Enunciado:
Você está desenvolvendo a lógica de um carrinho de compras para uma escola de programação. O sistema deve calcular o valor final da compra baseando-se em três critérios: quantidade de cursos, cupom de desconto e o perfil do usuário.

Regras de Negócio:

Desconto por Volume (Combo): Se o cliente comprar 3 ou mais cursos, ele recebe 30% de desconto automaticamente. Esta regra é prioritária e anula qualquer outra.

Cupom de Boas-vindas: Se o cliente usar o cupom "NEWUSER", ele ganha 50% de desconto, mas apenas se atender a duas condições:

Comprar exatamente 1 curso.

Ser um usuário novo no sistema (is_new_user deve ser True).

Validação de Erro: Se o usuário tentar usar o cupom "NEWUSER" sendo um cliente antigo ou comprando mais de um curso, o sistema deve avisar: "Cupom inválido para o seu perfil ou quantidade de itens".

Preço Base: Cada curso custa R$ 200,00.

```python
# Simulação de dados vindos do Banco de Dados
is_new_user = True

# Entradas do Usuário
course_quantity = int(input("Quantidade de cursos: "))
coupon_code = input("Digite seu cupom: ").upper().strip()

# Configurações do Sistema
course_unit_price = 200.0
discount_percent = 0.0

# 1. Verificação de Prioridade: Promoção Combo (3+)
if course_quantity >= 3:
    discount_percent = 0.30
    print("Sucesso: Combo aplicado (30% de desconto)!")

# 2. Verificação de Cupom de Boas-vindas
elif coupon_code == "NEWUSER":
    # Verificação das duas condições obrigatórias
    if course_quantity == 1 and is_new_user == True:
        discount_percent = 0.50
        print("Sucesso: Cupom NEWUSER aplicado (50% de desconto)!")
    else:
        print("Aviso: Cupom NEWUSER inválido para o seu perfil ou quantidade de itens.")

# 3. Caso padrão (Sem cupom ou cupom comum não listado)
else:
    print(f"Sem descontos aplicados.")

# Cálculos Finais
subtotal = course_quantity * course_unit_price
total_discount = subtotal * discount_percent
final_price = subtotal - total_discount

# Exibição do Recibo
print("-" * 35)
print(f"Cursos no carrinho: {course_quantity}")
print(f"Subtotal: R$ {subtotal:.2f}")
print(f"Desconto aplicado: R$ {total_discount:.2f}")
print(f"TOTAL A PAGAR: R$ {final_price:.2f}")
```
